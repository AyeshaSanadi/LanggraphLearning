```python
from langgraph.graph import START, END, StateGraph
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
class SimpleChatState(TypedDict):
    user_input: str
    bot_response: str
```

```python
def processing(state: SimpleChatState) -> SimpleChatState:
    user_input = state['user_input']
    
    prompt = f"Answer the following question: {user_input}"
    
    response = model.invoke(prompt)
    
    state['bot_response'] = response['choices'][0]['message']['content']
    
    return state
```

```python
graph = StateGraph(SimpleChatState)

graph.add_node("output_generate", processing)

graph.add_edge(START, "output_generate")
graph.add_edge("output_generate", END)

workflow = graph.compile()
```

```python
workflow.invoke({"user_input": "What is difference between cpu and vcpu?"})
```
